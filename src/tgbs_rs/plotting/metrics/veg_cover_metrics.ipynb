```python
import pandas as pd
from pathlib import Path

from tgbs_rs.config.config import TABLES_DIR
from tgbs_rs.plotting.table_prep import subset_metric_tables
from tgbs_rs.plotting.annual_metrics import (
    run_annual_metrics_workflow,
    plot_annual_metrics_core_figures,
    plot_annual_metrics_category_figures
)
from tgbs_rs.plotting.seasonal_metrics import (
    run_single_season_metrics_workflow,
    plot_single_season_metrics_core_figures
)
```

```python
csv_paths = {
    "hls_annual_master": TABLES_DIR / "tgbs_hls_annual_master_all_bands_2014_2025.csv",
    "hls_seasonal_master": TABLES_DIR / "tgbs_hls_seasonal_master_all_bands_2014_2025.csv",
    "s2_annual_master": TABLES_DIR / "tgbs_s2_annual_master_all_bands_2019_2025.csv",
    "s2_seasonal_master": TABLES_DIR / "tgbs_s2_seasonal_master_all_bands_2019_2025.csv",
}

hls_annual_master = pd.read_csv(csv_paths["hls_annual_master"])
hls_seasonal_master = pd.read_csv(csv_paths["hls_seasonal_master"])
s2_annual_master = pd.read_csv(csv_paths["s2_annual_master"])
s2_seasonal_master = pd.read_csv(csv_paths["s2_seasonal_master"])

# Define metric indices
VEG_COVER_METRICS = ["NDVI", "SAVI", "EVI"]
```

##### Generate HLS and Sentinel-2 Annual and Seasonal Subsets

```python
# Generate HLS annual and seasonal subsets
hls_annual_veg_cover, hls_seasonal_veg_cover = subset_metric_tables(
    annual_master=hls_annual_master,
    seasonal_master=hls_seasonal_master,
    metric_cols=VEG_COVER_METRICS,
)

# Generate Sentinel-2 annual and seasonal subsets
s2_annual_veg_cover, s2_seasonal_veg_cover = subset_metric_tables(
    annual_master=s2_annual_master,
    seasonal_master=s2_seasonal_master,
    metric_cols=VEG_COVER_METRICS,
)
```

##### Plot Annual Veg Cover Focal vs. Reference vs. Degraded

```python
hls_annual_cover_outputs = run_annual_metrics_workflow(
    annual_df=hls_annual_veg_cover,
    metric_cols=VEG_COVER_METRICS,
)
```

```python
# annual focal vs reference
plot_annual_metrics_core_figures(
    outputs=hls_annual_cover_outputs,
    source_label="HLS",
    metric_cols=VEG_COVER_METRICS,
    comparison_label="Focal Vs Reference",
    envelope_label="reference",
)
```

```python
# annual focal vs degraded
plot_annual_metrics_core_figures(
    outputs=hls_annual_cover_outputs,
    source_label="HLS",
    metric_cols=["NDVI", "SAVI"],
    comparison_label="Focal Vs Degraded",
    envelope_label="degraded",
)
```

##### Plot Annual Wet vs. Dry Season Veg Cover

```python
# wet seasonal
hls_wet_cover_outputs = run_single_season_metrics_workflow(
    seasonal_df=hls_seasonal_veg_cover,
    season="wet",
    metric_cols=VEG_COVER_METRICS,
)
```

```python
plot_single_season_metrics_core_figures(
    outputs=hls_wet_cover_outputs,
    season="wet",
    source_label="HLS",
    metric_cols=VEG_COVER_METRICS,
    comparison_label="Focal Vs Reference",
    envelope_label="reference",
)
```

```python
# dry seasonal
hls_dry_cover_outputs = run_single_season_metrics_workflow(
    seasonal_df=hls_seasonal_veg_cover,
    season="dry",
    metric_cols=VEG_COVER_METRICS,
)
```

```python
plot_single_season_metrics_core_figures(
    outputs=hls_dry_cover_outputs,
    season="dry",
    source_label="HLS",
    metric_cols=VEG_COVER_METRICS,
    comparison_label="Focal Vs Reference",
    envelope_label="reference",
)
```

##### Plot Sentinel-2 Annual Veg Cover Metrics

```python
# Sentinel-2 support
s2_annual_outputs = run_annual_metrics_workflow(
    annual_df=s2_annual_veg_cover,
    metric_cols=VEG_COVER_METRICS,
)
```

```python
plot_annual_metrics_category_figures(
    outputs=s2_annual_outputs,
    source_label="Sentinel-2",
    metric_cols=VEG_COVER_METRICS
)
```
