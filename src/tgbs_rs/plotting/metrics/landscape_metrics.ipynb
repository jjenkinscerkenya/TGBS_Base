```python
import pandas as pd
import matplotlib.pyplot as plt

from tgbs_rs.config.config import TABLES_DIR
from tgbs_rs.plotting.landscape_metrics import (
    load_metrics_table, 
    plot_landscape_metric_facets,
    plot_site_year_heatmap,
    plot_pland_trajectories,
    plot_pca_site_years,
)   
```

```python
land_df = load_metrics_table(TABLES_DIR / "landscape_metrics_landscape_level.csv")
class_df = load_metrics_table(TABLES_DIR / "landscape_metrics_class_level.csv")
land_df_all = load_metrics_table(TABLES_DIR / "landscape_metrics_all.csv")
```

```python
core_metrics = ["np", "pd", "ed", "lpi", "shdi", "ai"]
```

##### Faceted landscape trajectories

```python
g1 = plot_landscape_metric_facets(
    land_df,
    metrics=core_metrics,
    site_col="site_id",
)
plt.show()
```

##### Standardized site-year heatmap

```python
fig, ax = plot_site_year_heatmap(
    land_df,
    metrics=core_metrics,
    site_col="site_id",
    figsize=(10, 8),
)
plt.show()
```

##### Class-level pland trajectories

```python
g3 = plot_pland_trajectories(
    class_df,
    site_col="site_id",
    class_col="class",
)
plt.show()
```

##### PCA of site-years

```python
fig, ax = plot_pca_site_years(
    land_df,
    metrics=core_metrics,
    site_col="site_id",
    annotate=False,
)
plt.show()
```
