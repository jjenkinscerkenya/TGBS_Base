```python
from pathlib import Path
import pandas as pd

from tgbs_rs.config.config import TABLES_DIR
```

#### Basic shape, columns, dtypes

```python
csv_paths = {
    "hls_annual": TABLES_DIR / "tgbs_hls_annual_composite_all_bands_2014_2025_full_raw.csv",
    "hls_monthly": TABLES_DIR / "tgbs_hls_monthly_composite_all_bands_2014_2025_full_raw.csv",
    "s2_annual": TABLES_DIR / "tgbs_s2_annual_composite_all_bands_2019_2025_full_raw.csv",
    "s2_monthly": TABLES_DIR / "tgbs_s2_monthly_composite_all_bands_2019_2025_full_raw.csv",
}
```

```python
for name, path in csv_paths.items():
    df = pd.read_csv(path)
    print(f"\n--- {name} ---")
    print("shape:", df.shape)
    print("columns:", list(df.columns))
    print("\ndtypes:")
    print(df.dtypes.sort_index())
```

#### Quick preview and null counts

```python
for name, path in csv_paths.items():
    df = pd.read_csv(path)
    print(f"\n--- {name} preview ---")
    print(df.head(3))
    print("\nTop null counts:")
    print(df.isna().sum().sort_values(ascending=False).head(20))
```

#### Unique values in key metadata columns

```python
key_cols = [
    "site_id",
    "site_name",
    "site_category",
    "year",
    "month",
    "date",
    "temporal_scale",
    "composite_stat",
    "image_count",
]

for name, path in csv_paths.items():
    df = pd.read_csv(path)
    print(f"\n--- {name} key metadata summary ---")
    for col in key_cols:
        if col in df.columns:
            print(f"\n{col}:")
            print(df[col].dropna().unique()[:20])
            print("n_unique:", df[col].nunique(dropna=True))
```

#### Candidate numeric data columns

```python
for name, path in csv_paths.items():
    df = pd.read_csv(path)
    numeric_cols = df.select_dtypes(include="number").columns.tolist()
    print(f"\n--- {name} numeric columns ---")
    print(numeric_cols)
```

#### Numeric summary for priority indices

```python
priority_cols = ["NIRv", "EVI", "NDVI", "BLUE", "GREEN", "RED", "NIR", "SWIR1", "SWIR2"]

for name, path in csv_paths.items():
    df = pd.read_csv(path)
    existing = [col for col in priority_cols if col in df.columns]
    print(f"\n--- {name} priority summaries ---")
    if existing:
        print(df[existing].describe().T)
```

#### Check row uniqueness for expected grain

```python
grain_checks = {
    "annual": ["site_id", "year"],
    "monthly": ["site_id", "year", "month"],
}

for name, path in csv_paths.items():
    df = pd.read_csv(path)
    grain = grain_checks["monthly"] if "monthly" in name else grain_checks["annual"]
    existing = [c for c in grain if c in df.columns]
    print(f"\n--- {name} grain check ---")
    print("expected grain:", existing)
    print("duplicate rows at grain:", df.duplicated(subset=existing).sum())
```
