```python
import ee
import geemap
```

```python
from tgbs_rs.config.config import GEE_PROJECT
```

```python
# Authenticate and Initialize Earth Engine
ee.Authenticate()
ee.Initialize(project= GEE_PROJECT)
```

```python
#Initialize Map
Map = geemap.Map()
Map = geemap.Map(height="800px")
```

```python
# Kenya boundary from a public country boundaries dataset.
kenya = (
    ee.FeatureCollection("USDOS/LSIB_SIMPLE/2017")
    .filter(ee.Filter.eq("country_na", "Kenya"))
)

# WDPA protected area polygons filtered to Kenya.
kenya_pas = (
    ee.FeatureCollection("WCMC/WDPA/current/polygons")
    .filterBounds(kenya.geometry())
)
```

```python
shimba_feature = ee.Feature(kenya_pas.filter(ee.Filter.eq("NAME", "Shimba")).first())

shimba_parts = kenya_pas.filter(
    ee.Filter.inList("NAME", ["Shimba", "Mkongani West", "Mkongani North"])
)

shimba_hills = ee.FeatureCollection([
    ee.Feature(shimba_parts.geometry().dissolve())
    .copyProperties(shimba_feature)
    .set("NAME", "Shimba Hills")
])

buda = kenya_pas.filter(ee.Filter.eq("NAME", "Buda"))

gogoni = kenya_pas.filter(ee.Filter.eq("NAME", "Gogoni"))
```
