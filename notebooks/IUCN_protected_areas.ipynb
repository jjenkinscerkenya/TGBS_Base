```python
import ee
import geemap

# Authenticate and Initialize Earth Engine
ee.Authenticate()
ee.Initialize(project="charrell-personal") 
```

```python
#Initialize Map
Map = geemap.Map(height="800px")
Map.add_basemap = 'SATELLITE'
```

```python
# Global Administrative Boundaries and IUCN Protected Areas
protected_areas = ee.FeatureCollection("WCMC/WDPA/current/polygons")

# Select protected areas in Kenya
kenya_pas = protected_areas.filter(ee.Filter.eq('ISO3', 'KEN'))
```

```python
shimba_hills = kenya_pas.filter(ee.Filter.eq('NAME', 'Shimba Hills'))
mkongani_north = kenya_pas.filter(ee.Filter.eq('NAME', 'Mkongani North'))
mkongani_west = kenya_pas.filter(ee.Filter.eq('NAME', 'Mkongani West'))

shimba = shimba_hills.merge(mkongani_north).merge(mkongani_west)

buda = kenya_pas.filter(ee.Filter.eq('NAME', 'Buda'))

gogoni = kenya_pas.filter(ee.Filter.eq('NAME', 'Gogoni'))
```

```python
pa_style = {
    'color': 'red',
    'fillColor': '00000000',
    'width': 2
}
```

```python
Map.addLayer(shimba, pa_style, 'Shimba Complex', False)
Map.addLayer(buda, pa_style, 'Buda', False)
Map.addLayer(gogoni, pa_style, 'Gogoni', False)

Map
```
