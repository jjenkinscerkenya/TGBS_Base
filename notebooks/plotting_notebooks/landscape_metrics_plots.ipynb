```python
import pandas as pd
import matplotlib.pyplot as plt

from tgbs_rs.config.config import TABLES_DIR
from tgbs_rs.metrics import (
    drop_site_year,
    load_metrics_table, 
    plot_landscape_metric_facets,
    plot_site_year_heatmap,
    plot_pland_trajectories,
    plot_pca_site_years,
    plot_connectivity_metric_facets,
    plot_core_metric_facets,
    plot_shape_metric_trajectories,
    plot_area_metric_trajectories,
    plot_metric_heatmap,
    plot_metric_small_multiples
)   
```

```python
land_df = load_metrics_table(TABLES_DIR / "landscape_metrics_landscape_level.csv")
class_df = load_metrics_table(TABLES_DIR / "landscape_metrics_class_level.csv")
```

The shimba_hills 2022 site-year raster was excluded from the primary landscape-metrics analysis because valid pixel coverage was approximately 55%, substantially lower than the other site-years, which were near complete coverage. This level of masking was considered likely to distort patch configuration and fragmentation metrics by introducing artificial gaps and patch breaks, so the site-year was omitted under the study’s raster quality-control rule.

```python
land_df = drop_site_year(land_df, "shimba_hills", 2022)
class_df = drop_site_year(class_df, "shimba_hills", 2022)
```

```python
core_metrics = ["np", "pd", "ed", "lpi", "shdi", "ai"]
```

##### Faceted landscape trajectories

```python
g1 = plot_landscape_metric_facets(
    land_df,
    metrics=core_metrics,
    site_col="site_id",
)
plt.show()
```

##### Standardized site-year heatmap

```python
fig, ax = plot_site_year_heatmap(
    land_df,
    metrics=core_metrics,
    site_col="site_id",
    figsize=(10, 8),
)
plt.show()
```

##### Class-level pland trajectories

```python
g3 = plot_pland_trajectories(
    class_df,
    site_col="site_id",
    class_col="class",
)
plt.show()
```

##### PCA of site-years

```python
fig, ax = plot_pca_site_years(
    land_df,
    metrics=core_metrics,
    site_col="site_id",
    annotate=False,
)
plt.show()
```

##### Core area: CORE_MN, CPLAND, TCA

```python
g = plot_core_metric_facets(
    class_df,
    site_col="site_id",
    metrics=["core_mn", "cpland", "tca"],
    level="class",
    class_value=1,
)
plt.show()
```

##### Shape: SHAPE_MN

```python
fig, ax = plot_shape_metric_trajectories(
    class_df,
    site_col="site_id",
    level="class",
    class_value=1,
)
plt.show()
```

##### Area: AREA_MN

```python
fig, ax = plot_area_metric_trajectories(
    class_df,
    site_col="site_id",
    level="class",
    class_value=1,
)
plt.show()
```

##### Heatmap for a single recovery-relevant metric

```python
fig, ax = plot_metric_heatmap(
    class_df,
    metric="cpland",
    site_col="site_id",
    level="class",
    class_value=1,
)
plt.show()
```
