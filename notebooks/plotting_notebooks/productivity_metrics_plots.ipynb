```python
import pandas as pd
from pathlib import Path

from tgbs_rs.config.config import TABLES_DIR
from tgbs_rs.visualization.tables import subset_metric_tables
from tgbs_rs.metrics.temporal import (
    run_annual_metrics_workflow,
    plot_annual_metrics_core_figures,
    plot_annual_metrics_category_figures,
    run_single_season_metrics_workflow,
    plot_single_season_metrics_core_figures
)
```

```python
csv_paths = {
    "hls_annual_master": TABLES_DIR / "tgbs_hls_annual_master_all_bands_2014_2025.csv",
    "hls_seasonal_master": TABLES_DIR / "tgbs_hls_seasonal_master_all_bands_2014_2025.csv",
    "s2_annual_master": TABLES_DIR / "tgbs_s2_annual_master_all_bands_2019_2025.csv",
    "s2_seasonal_master": TABLES_DIR / "tgbs_s2_seasonal_master_all_bands_2019_2025.csv",
}

hls_annual_master = pd.read_csv(csv_paths["hls_annual_master"])
hls_seasonal_master = pd.read_csv(csv_paths["hls_seasonal_master"])
s2_annual_master = pd.read_csv(csv_paths["s2_annual_master"])
s2_seasonal_master = pd.read_csv(csv_paths["s2_seasonal_master"])

# Define metric indices
PRODUCTIVITY_METRICS = ["NIRv", "EVI", "NDVI"]
```

##### Generate HLS and Sentinel-2 Annual and Seasonal Productivity Subsets

```python
# Generate HLS annual and seasonal subsets
hls_annual_productivity, hls_seasonal_productivity = subset_metric_tables(
    annual_master=hls_annual_master,
    seasonal_master=hls_seasonal_master,
    metric_cols=PRODUCTIVITY_METRICS,
)

# Generate Sentinel-2 annual and seasonal subsets
s2_annual_productivity, s2_seasonal_productivity = subset_metric_tables(
    annual_master=s2_annual_master,
    seasonal_master=s2_seasonal_master,
    metric_cols=PRODUCTIVITY_METRICS,
)
```

##### Plot Annual Productivity Focal vs. Reference vs. Degraded

```python
hls_annual_outputs = run_annual_metrics_workflow(
    annual_df=hls_annual_productivity,
    metric_cols=PRODUCTIVITY_METRICS,
)
```

```python
# HLS annual focal vs reference
plot_annual_metrics_core_figures(
    outputs=hls_annual_outputs,
    source_label="HLS",
    metric_cols=PRODUCTIVITY_METRICS,
    comparison_label="Focal vs Bio Corridor vs Reference",
    envelope_label="reference",
)
```

```python
plot_annual_metrics_category_figures(
    outputs=hls_annual_outputs,
    source_label="HLS",
    metric_cols=PRODUCTIVITY_METRICS
)
```

```python
# HLS annual focal vs degraded
plot_annual_metrics_core_figures(
    outputs=hls_annual_outputs,
    source_label="HLS",
    metric_cols=["NIRv"],
    comparison_label="Focal vs Bio Corridor vs Degraded",
    envelope_label="degraded",
)
```

##### Plot Annual Wet vs. Dry Season Productivity

```python
# seasonal wet
hls_wet_outputs = run_single_season_metrics_workflow(
    seasonal_df=hls_seasonal_productivity,
    season="wet",
    metric_cols=["NIRv", "EVI"],
)
```

```python
plot_single_season_metrics_core_figures(
    outputs=hls_wet_outputs,
    season="wet",
    source_label="HLS",
    metric_cols=["NIRv", "EVI"],
    comparison_label="Focal vs Bio Corridor vs Reference",
    envelope_label="reference",
)
```

```python
# seasonal dry
hls_dry_outputs = run_single_season_metrics_workflow(
    seasonal_df=hls_seasonal_productivity,
    season="dry",
    metric_cols=["NIRv", "EVI"],
)
```

```python
plot_single_season_metrics_core_figures(
    outputs=hls_dry_outputs,
    season="dry",
    source_label="HLS",
    metric_cols=["NIRv", "EVI"],
    comparison_label="Focal vs Bio Corridor vs Reference",
    envelope_label="reference",
)
```

##### Plot Sentinel-2 Annual Productivity Metrics

```python
# Sentinel-2 support
s2_annual_outputs = run_annual_metrics_workflow(
    annual_df=s2_annual_productivity,
    metric_cols=PRODUCTIVITY_METRICS,
)
```

```python
plot_annual_metrics_core_figures(
    outputs=s2_annual_outputs,
    source_label="Sentinel-2",
    metric_cols=PRODUCTIVITY_METRICS
)
```

```python
plot_annual_metrics_category_figures(
    outputs=s2_annual_outputs,
    source_label="Sentinel-2",
    metric_cols=PRODUCTIVITY_METRICS
)
```
